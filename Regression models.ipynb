```python
import numpy as np
import matplotlib.pyplot as plt

#Create dataset for linear regression
x_data = np.linspace(1,100,1000)
y_data = 2.34*x_data + 1.67 + np.random.normal(0,15,1000)
plt.scatter(x_data,y_data,s=1) 
plt.title("Data-set")
```

```python
#general model for n-degree loss function
def custom_sum1(arr_y,arr_x,degree):
    sum=0
    for i in range(arr_x.size):
        if(arr_y[i]>0):
            sum-=arr_x[i]*(arr_y[i]**(degree-1))
        else :
            sum+=arr_x[i]*(arr_y[i]**(degree-1))
    return sum

def custom_sum2(arr_y,degree):
    sum=0
    for i in range(arr_y.size):
        if(arr_y[i]>0):
            sum-=(arr_y[i]**(degree-1))
        else :
            sum+=(arr_y[i]**(degree-1))
    return sum

def training(x,y,iteration,learning_val,degree):
    m=0
    c=0
    n=x.size

    for i in range(iteration):
        y_pre = m*x + c
        D_m = (degree/n)*custom_sum1(y-y_pre,x,degree)
        D_c = (degree/n)*custom_sum2(y-y_pre,degree)
        m = m - learning_val*D_m
        c = c - learning_val*D_c

    print(m,c)
    plt.scatter(x,y,s=1,label="Data set")
    plt.plot(x,m*x+c,color="red",label="Optimized function")
    plt.title("Optimization with degree {} loss function".format(degree))
    plt.legend()
```

```python
training(x_data,y_data,1000,0.000001,3)         #degree 3 loss function
```

```python
training(x_data,y_data,1000,0.001,1)           #degree 1 loss function
```

```python
#create data set for polynomial regression
x_data = np.linspace(1,100,1000)
y_data = 2*(x_data)**2 + 3*x_data +8 + np.random.normal(0,1000,1000)
plt.scatter(x_data,y_data,s=1)
plt.title("Data-set")
```

```python
#general model for n-degree loss function on a quadratic data points
def custom_sum_a(arr_y,arr_x,degree):
    sum=0
    for i in range(arr_x.size):
        if(arr_y[i]>0):
            sum-=(arr_x[i]**2)*(arr_y[i]**(degree-1))
        else :
            sum+=(arr_x[i]**2)*(arr_y[i]**(degree-1))
    return sum

def custom_sum_b(arr_y,arr_x,degree):
    sum=0
    for i in range(arr_x.size):
        if(arr_y[i]>0):
            sum-=arr_x[i]*(arr_y[i]**(degree-1))
        else :
            sum+=arr_x[i]*(arr_y[i]**(degree-1))
    return sum

def custom_sum_c(arr_y,degree):
    sum=0
    for i in range(arr_y.size):
        if(arr_y[i]>0):
            sum-=(arr_y[i]**(degree-1))
        else :
            sum+=(arr_y[i]**(degree-1))
    return sum

def training(x,y,iteration,learning_val,degree):
    a=0                         #error function ax^2 + bx + c
    b=0
    c=0
    n=x.size

    for i in range(iteration):
        y_pre = a*(x**2)+ b*x + c
        D_a = (degree/n)*custom_sum_a(y-y_pre,x,degree)
        D_b = (degree/n)*custom_sum_b(y-y_pre,x,degree)
        D_c = (degree/n)*custom_sum_c(y-y_pre,degree)
        a = a - learning_val*D_a
        b = b - learning_val*D_b
        c = c - learning_val*D_c

    print(a,b,c)
    plt.scatter(x,y,s=1,label="Data set")
    plt.plot(x,a*(x**2)+ b*x + c,color="red",label="Optimized function")
    plt.title("Optimization with degree {} loss function".format(degree))
    plt.legend()
```

```python
training(x_data,y_data,5000,10**-18,4)          #degree 4 loss function
```

```python
training(x_data,y_data,11000,10**-29,7)        #degree 7 loss function
```
